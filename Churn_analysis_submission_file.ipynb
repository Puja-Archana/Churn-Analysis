###  Separating Training Dataset to x_train and y_train

```python
x_train = train_ds.drop("Is_Churn", axis = 'columns')
```

```python
# Print all the columns and its unique values
for column in x_train:
    print(f'{column} : {x_train[column].unique()}')
```

```python
x_train["Gender"] = x_train["Gender"].replace(['Male','Female'],[1,0])
```

```python
# Here ID is of object type, which is not required in for our modeling. So we drop the column.
x_train.drop('ID', axis ='columns', inplace = True)
x_train.dtypes
```

### One Hot Encoding

```python
# OneHotEncoding of following columns
x_train1 = pd.get_dummies(data = x_train, columns = ['Income','Vintage','Product_Holdings','Credit_Category'])
```

```python
x_train1
```

```python
x_train1
```

```python
# Printing all columns and its unique values

for column in x_train1:
    print(f'{column} : {x_train1[column].unique()}')
```

### Apply feature scaling

```python
# Applying MinMaxScaler to columns
cols_to_scale = ['Age','Balance']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train1[cols_to_scale] = scaler.fit_transform(x_train1[cols_to_scale])
```

```python
# print all columns and its unique values

for cols in x_train1:
    print(f'{cols} : {x_train1[cols].unique()}')
```
